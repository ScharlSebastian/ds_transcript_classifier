```python
from youtube_transcript_api import YouTubeTranscriptApi
from operator import itemgetter 
```

```python
def get_transcript_from_yt(url):
    
    try:
        yt_video_id = url.split("?v=")[1]
        resp = YouTubeTranscriptApi.get_transcript(yt_video_id)
        data_list = list(map(itemgetter('text'), resp))
        data = data_list
    except:
        data = "Error"
    
    return data
```

```python
url = "https://www.youtube.com/watch?v=fB7nyxXaczY"

transcripted_data = get_transcript_from_yt(url)
transcripted_data
```
